```python
from pyecharts.charts import Boxplot
from pyecharts.faker import Faker
import seaborn as sns
from sklearn import datasets
import pandas as pd
from pyecharts import options as opts
```

```python
athletes= pd.read_csv("F:/05seaborn/athlete_events.csv")
```

```python
countrys = {
    'CHN':"中国",
    'JPN':"日本",
    "KOR":"韩国",
    'USA':"美国",
    'CAV':"加拿大",
    'BRA':"巴西",
    'GBR':"英国",
    'FRA':"法国",
    'ITA':"意大利",
    'ETH':"埃塞俄比亚",
    'KEN':"肯尼亚",
    'NTG':"尼日利亚"
}
```

```python
my_athletes = athletes[athletes['NOC'].isin(list(countrys.keys()))]
man = []
woman =[]
for country in countrys.keys():
    m_aths = my_athletes[(my_athletes["NOC"]==country) & (my_athletes["Sex"]=="M")]["Height"].dropna()
    f_aths = my_athletes[(my_athletes["NOC"]==country) & (my_athletes["Sex"]=="F")]["Height"].dropna()
    man.append(m_aths)
    woman.append(f_aths)
```

```python
c = Boxplot(init_opts=opts.InitOpts(bg_color="#ccc"))
c.add_xaxis(list(countrys.values()))
c.add_yaxis("男性", c.prepare_data(man))
c.add_yaxis("女性", c.prepare_data(woman))
c.set_global_opts(yaxis_opts=opts.AxisOpts(
    name_rotate= -15,
    min_=130, max_=230,
    interval=20, #y 间隔
    name_location="center"
))
c.render_notebook()
```
